```python


pd.read_csv(): Read data from a CSV file.
pd.read_excel(): Read data from an Excel file.
pd.read_sql(): Read data from a SQL database.
pd.read_json(): Read data from a JSON file.
Exploring Data:

df.head(), df.tail(): Display the first/last n rows of a DataFrame.
df.info(): Display a concise summary of the DataFrame.
df.describe(): Generate descriptive statistics.
df.shape: Returns the dimensions of the DataFrame.
df.columns: Returns the column labels.
df.dtypes: Returns the data types of each column.
Selection and Indexing:

df['column_name']:
df[['col1', 'col2']]:
df.loc[]:
df.iloc[]:

df[df['column'] > value]:
df.query():
Handling Missing Data:

df.isnull(), df.notnull():
df.dropna():
df.fillna():
Manipulating Data:

df.groupby():
df.sort_values():
df.pivot_table():
df.melt():
Combining DataFrames:

pd.concat():
pd.merge():
Changing Data:

df.apply():
df.map():
Time Series Analysis:

pd.to_datetime():
df.resample():
df.shift():
Statistical Functions:

df.mean(), df.median(), df.std():
Data Cleaning:

df.rename():
df.drop():
File Output:

df.to_csv(): Write DataFrame to a CSV file.
df.to_excel(): Write DataFrame to an Excel file.
```
